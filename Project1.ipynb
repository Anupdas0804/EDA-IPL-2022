```python
bowling = iplB.groupby(['bowler']).agg({'overs':'count','total_run':'sum'})
bowling['overs']=bowling['overs']/6
bowling['overs']=bowling['overs'].astype('int')
bowling.sort_values(by='total_run', ascending = False)
bowling['Economy']=bowling['total_run']/bowling['overs']
bowling.sort_values(by='Economy', ascending = False)
```

```python
bowling[bowling['overs']!=0].sort_values(by='Economy', ascending = False).head(10)
```

```python
bowling[bowling['overs']!=0].sort_values(by='Economy', ascending = False).head(10).plot.bar(figsize=(15,6), title = 'overs vs. total_run vs. Economy')
```
